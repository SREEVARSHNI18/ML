```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import seaborn as sns
x=np.array([14.2,16.4,11.9,15.2,18.5,22.1,19.4,25.1,23.4,18.1,22.6,17.2]).reshape([-1,1])
y=np.array([215,325,185,332,406,522,412,614,544,421,445,408])
model=LinearRegression()
model.fit(x,y)
scr=model.score(x,y)
print("Score of the model: ",scr)
intercept=model.intercept_ #b0
slope=model.coef_[0] #b1
print("Intercept of the Equation: ",intercept)
print("slope: ",slope)
print(f"The linear regression of the data is: y={intercept:.2f}+{slope:.2f}x")

plt.scatter(x,y,color='blue',label='Data Points')
plt.title("Ice cream Sales Vs Temprature")
plt.xlabel("temprature ")
plt.ylabel("ice cream sales")
plt.legend()

```

```python
temp_interpolate = np.array([[21]])
temp_extrapolate = np.array([[29]])
sales_interpolate = model.predict(temp_interpolate)
sales_extrapolate = model.predict(temp_extrapolate)

print(f"Interpolated sales value at 21째C: ${sales_interpolate[0]:.2f}")
print(f"Extrapolated sales value at 29째C: ${sales_extrapolate[0]:.2f}")
```
