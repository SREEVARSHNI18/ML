```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#using np
x=np.array([14.2,16.4,11.9,15.2,18.5,22.1,19.4,25.1,23.4,18.1,22.6,17.2])
y=np.array([215,325,185,332,406,522,412,614,544,421,445,408])
r=np.corrcoef(x,y)
print("correlation: \n",r)

#using pandas
df=pd.DataFrame({'x':x,'y':y})
#correlation = x.corr(y)  # calculate the correlation
#print("correlation: ", correlation)

```

```python
print(len(x))
print(len(y))
X = np.vstack([x, np.ones(len(x))]).T
b = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print("Line of best fit: y = {:.2f}x + {:.2f}".format(b[0], b[1]))
```

```python
interpolated_sales = b[0] * 21 + b[1]
print("Interpolated sales at 21°: {:.2f}".format(interpolated_sales))

# extrapolate a sales value at 29°
extrapolated_sales = b[0] * 29 + b[1]
print("Extrapolated sales at 29°: {:.2f}".format(extrapolated_sales))

plt.scatter(x,y)
plt.plot(x, b[0] * x + b[1], 'r', label='Line of best fit')
plt.xlabel("degree Celcius")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
