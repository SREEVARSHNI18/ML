```python
import pandas as pd
df=pd.read_csv("IPL.csv")
#print(df.describe())
#Perform joining of data frame- compare the average SOLD PRICE for different AGE categories with the different age and PLAYING ROLE categories, we need to merge the DataFrames soldprice_by_age and soldprice_by_ age_role. The common column is AGE and this needs outer join. Age is considered as categorical variable(as 1,2,3)
sold_price_age=df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
sold_price_age.columns=['AGE','AGE_SOLD_PRICE']

sold_price_agerole=df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
sold_price_agerole.columns=['AGE','PLAYING ROLE','AGEROLE_SOLDPRICE']

merged_df=pd.merge(sold_price_age,sold_price_agerole,on='AGE',how='outer')
print(merged_df)
```

```python
sold_price_agerole=df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
sold_price_agerole.columns=['AGE','PLAYING ROLE','AVG_SOLD_PRICEROLE']

sold_price_age=df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
sold_price_age.columns=['AGE','AVG_SOLD_PRICE']
merge_data=pd.merge(sold_price_age,sold_price_agerole,on='AGE',how='outer')
merge_data['premium %']=((merge_data['AVG_SOLD_PRICEROLE']-merge_data['AVG_SOLD_PRICE'])/merge_data['AVG_SOLD_PRICE'])*100
print(merge_data)
```
