```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
df=pd.read_csv(r"C:\Users\sreev\OneDrive\Desktop\college\sem5\ML\ps2\advertising.csv")
print(df.head())
#checking for number of null values
print(df.isnull().sum())
```

```python
x=df[['Sales']]
y=df['TV']
model=LinearRegression()
model.fit(x,y)
scr=model.score(x,y)
print("R^2 score of the model:", scr)
intercept=model.intercept_
slope=model.coef_[0]
print(f"The linear regression equation is: y={intercept:.2f}+{slope:.2f}x")
plt.scatter(x,y,color="red",label="Data points")
plt.plot(x, model.predict(x), color='blue', linewidth=3, label="Linear regression line")
plt.xlabel("Sales")
plt.ylabel("TV")
plt.title("Relationship between TV and Sales")
plt.show()

```

```python
y_pred=model.predict(x)
residuals=y-y_pred
n=len(y_pred)
mse=np.mean(residuals**2)
var_x = np.var(x.values.flatten(), ddof=1)  #variance
std_error_slope = np.sqrt(mse / ((n - 1) * var_x)) 

alpha = 0.05
t_critical = 2.093  # for a 95% confidence level with (n-2) degrees of freedom

# Calculate confidence interval for the slope
slope = model.coef_[0]
slope_standard_error = std_error_slope
slope_lower = slope - t_critical * slope_standard_error
slope_upper = slope + t_critical * slope_standard_error

print(f"95% Confidence Interval for Slope: ({slope_lower}, {slope_upper})")
```
